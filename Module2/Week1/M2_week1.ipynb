```python
!gdown 1iA0WmVfW88HyJvTBSQDI5vesf-pgKabq
import pandas as pd
df = pd.read_csv('/content/advertising.csv')
data = df.to_numpy()
print(data)
```

```python

column_4 = data[:, 3]
max_value = np.max(column_4)
print(max_value)
for index, value in enumerate(column_4):
    if value == max_value:
        max_index = index
        break
print(max_index)
mean_tv = np.mean(data[:, 0])
print(mean_tv)

banghi = np.sum(data[:,3]>=20)
print(banghi)

#cột sale >=15
filtered_data = data[data[:, 3] >= 15]
#tính giá trị trung bình dựa trên filtered
mean_sale = np.mean(filtered_data[:, 1])
print(mean_sale)

mean_newspaper = np.mean(data[:, 2])
filtered_data = data[data[:, 2] > mean_newspaper]
sum_sales = np.sum(filtered_data[:, 3])
print(sum_sales)
mean_sales = np.mean(data[:, 3])
scores = np.array(["Good" if value > mean_sales else "Bad" if value < mean_sales else "Average" for value in data[:, 3]])
print(scores[7:10])

mean_sales = np.mean(data[:, 3])
A = data[:, 3][np.abs(data[:, 3] - mean_sales).argmin()]
scores = np.array(["Good" if value > A else "Bad" if value < A else "Average" for value in data[:, 3]])
print(scores[7:10])
```
